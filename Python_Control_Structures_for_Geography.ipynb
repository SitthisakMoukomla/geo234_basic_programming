````python
# ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á: ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤‡∏ß‡∏±‡∏ô‡∏ô‡∏±‡πâ‡∏ô‡∏ù‡∏ô‡∏ï‡∏Å‡∏´‡∏£‡∏∑‡∏≠‡πÑ‡∏°‡πà
rain = 5

if rain > 0:
    print(f"üåßÔ∏è ‡∏ß‡∏±‡∏ô‡∏ô‡∏µ‡πâ‡∏ù‡∏ô‡∏ï‡∏Å {rain} mm")
else:
    print("‚òÄÔ∏è ‡∏ß‡∏±‡∏ô‡∏ô‡∏µ‡πâ‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏ù‡∏ô")
````

## 2.1 ‡∏ß‡∏ô‡∏ã‡πâ‡∏≥‡πÉ‡∏ô list

````python
# ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á: ‡πÅ‡∏™‡∏î‡∏á‡∏≠‡∏∏‡∏ì‡∏´‡∏†‡∏π‡∏°‡∏¥‡∏ó‡∏±‡πâ‡∏á 8 ‡∏ß‡∏±‡∏ô
print("‡∏≠‡∏∏‡∏ì‡∏´‡∏†‡∏π‡∏°‡∏¥‡∏£‡∏≤‡∏¢‡∏ß‡∏±‡∏ô:")
for temp in temperatures:
    print(f"  üå°Ô∏è {temp}¬∞C")
````

## 2.2 ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢

````python
# ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á: ‡∏Ñ‡∏≥‡∏ô‡∏ß‡∏ì‡∏Ñ‡πà‡∏≤‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢‡∏≠‡∏∏‡∏ì‡∏´‡∏†‡∏π‡∏°‡∏¥
total = 0
count = 0

for temp in temperatures:
    total += temp
    count += 1

average = total / count
print(f"‡∏≠‡∏∏‡∏ì‡∏´‡∏†‡∏π‡∏°‡∏¥‡πÄ‡∏â‡∏•‡∏µ‡πà‡∏¢: {average:.2f}¬∞C")
````

## 2.3 ‡∏ß‡∏ô‡∏ã‡πâ‡∏≥‡∏î‡πâ‡∏ß‡∏¢ range()

````python
# ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á: ‡πÅ‡∏™‡∏î‡∏á‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà 1-10
print("‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡πÉ‡∏ô‡πÄ‡∏î‡∏∑‡∏≠‡∏ô:")
for day in range(1, 11):  # 1-10
    print(f"  ‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà {day}")
````

## 2.4 ‡∏ß‡∏ô‡∏ã‡πâ‡∏≥‡∏û‡∏£‡πâ‡∏≠‡∏°‡πÄ‡∏á‡∏∑‡πà‡∏≠‡∏ô‡πÑ‡∏Ç

````python
# ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á: ‡∏ô‡∏±‡∏ö‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡∏ù‡∏ô‡∏ï‡∏Å
rainy_days = 0

for rain in rainfall:
    if rain > 0:
        rainy_days += 1

print(f"‡∏à‡∏≥‡∏ô‡∏ß‡∏ô‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡∏ù‡∏ô‡∏ï‡∏Å: {rainy_days} ‡∏ß‡∏±‡∏ô")
````

````python
# ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á: ‡∏´‡∏≤‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡∏≠‡∏∏‡∏ì‡∏´‡∏†‡∏π‡∏°‡∏¥‡∏™‡∏π‡∏á‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î
max_temp = temperatures[0]
max_day = 1

for i in range(len(temperatures)):
    if temperatures[i] > max_temp:
        max_temp = temperatures[i]
        max_day = i + 1

print(f"‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà {max_day} ‡∏°‡∏µ‡∏≠‡∏∏‡∏ì‡∏´‡∏†‡∏π‡∏°‡∏¥‡∏™‡∏π‡∏á‡∏ó‡∏µ‡πà‡∏™‡∏∏‡∏î {max_temp}¬∞C")
````

## 2.5 ‡πÉ‡∏ä‡πâ enumerate() - ‡πÅ‡∏™‡∏î‡∏á‡∏ó‡∏±‡πâ‡∏á‡∏•‡∏≥‡∏î‡∏±‡∏ö‡πÅ‡∏•‡∏∞‡∏Ñ‡πà‡∏≤

````python
# ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á: ‡πÅ‡∏™‡∏î‡∏á‡∏ß‡∏±‡∏ô‡πÅ‡∏•‡∏∞‡∏≠‡∏∏‡∏ì‡∏´‡∏†‡∏π‡∏°‡∏¥‡∏û‡∏£‡πâ‡∏≠‡∏°‡∏Å‡∏±‡∏ô
print("‡∏≠‡∏∏‡∏ì‡∏´‡∏†‡∏π‡∏°‡∏¥‡∏£‡∏≤‡∏¢‡∏ß‡∏±‡∏ô:")
for day, temp in enumerate(temperatures, start=1):
    print(f"  ‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà {day}: {temp}¬∞C")
````
